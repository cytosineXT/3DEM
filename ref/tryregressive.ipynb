```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# 数据
data = {
    'Scale': [1.3119, 1.2536, 1.1942, 1.1175, 1.2609, 1.2724, 1.2769, 1.1346, 1.1701, 1.2736],
    'Area': [0.5884, 0.4471, 0.3755, 0.3554, 0.5295, 0.4908, 0.4452, 0.6574, 0.2665, 0.6269],
    'Volume': [0.0101, 0.0067, 0.0050, 0.0053, 0.0077, 0.0076, 0.0066, 0.0098, 0.0025, 0.0103],
    '起点': [0.032, 0.097, 0.11, 0.1, 0.14, 0.105, 0.105, 0.076, 0.165, 0.12]
}

df = pd.DataFrame(data)

# 定义自变量和因变量
X = df[['Scale', 'Area', 'Volume']]
y = df['起点']

# 创建线性回归模型
linear_model = LinearRegression()
linear_model.fit(X, y)

# 线性回归预测
linear_pred = linear_model.predict(X)

# 创建多项式回归模型（二次项）
poly_features = PolynomialFeatures(degree=2, include_bias=False)
poly_model = make_pipeline(poly_features, LinearRegression())
poly_model.fit(X, y)

# 多项式回归预测
poly_pred = poly_model.predict(X)

# 提取多项式回归模型的系数
poly_coefficients = poly_model.named_steps['linearregression'].coef_
poly_intercept = poly_model.named_steps['linearregression'].intercept_

# 获取多项式特征名称
poly_feature_names = poly_features.get_feature_names_out(X.columns)

# 打印多项式回归公式
formula_terms = [f"{coef:.4f}*{name}" for coef, name in zip(poly_coefficients, poly_feature_names)]
poly_formula = " + ".join(formula_terms) + f" + {poly_intercept:.4f}"
print("Polynomial Regression Formula:")
print(f"起点 = {poly_formula}")


# 绘制图形
plt.figure(figsize=(10, 6))

# 原始数据点
plt.scatter(range(len(y)), y, color='blue', label='Original Data')

# 线性回归拟合曲线
plt.plot(range(len(y)), linear_pred, color='red', linestyle='-', label='Linear Fit')

# 多项式回归拟合曲线
plt.plot(range(len(y)), poly_pred, color='green', linestyle='--', label='Polynomial Fit')

# 图例和标签
plt.legend()
plt.xlabel('Index')
plt.ylabel('Mie Start Point')
plt.title('Linear vs Polynomial Fit')
plt.grid(True)

# 显示图形
plt.savefig('regre.png')
plt.show()
```
