```python
import torch
import torch.nn as nn

class YourModel(nn.Module):
    def __init__(self, decoder_outdim):
        super(YourModel, self).__init__()
        self.decoder_outdim = decoder_outdim
        
        # 定义网络层
        self.upconv1 = nn.ConvTranspose2d(decoder_outdim*8, int(decoder_outdim*8/2), kernel_size=2, stride=2)
        self.bn1 = nn.BatchNorm2d(int(decoder_outdim*8/2))
        self.conv1_1 = nn.Conv2d(int(decoder_outdim*8/2), int(decoder_outdim*8/2), kernel_size=3, stride=1, padding=1)
        self.conv1_2 = nn.Conv2d(int(decoder_outdim*8/2), int(decoder_outdim*8/2), kernel_size=3, stride=1, padding=1)
        self.bn1_1 = nn.BatchNorm2d(int(decoder_outdim*8/2))
        self.bn1_2 = nn.BatchNorm2d(int(decoder_outdim*8/2))
        
        self.upconv2 = nn.ConvTranspose2d(int(decoder_outdim*8/2), int(decoder_outdim*8/4), kernel_size=2, stride=2)
        self.bn2 = nn.BatchNorm2d(int(decoder_outdim*8/4))
        self.conv2_1 = nn.Conv2d(int(decoder_outdim*8/4), int(decoder_outdim*8/4), kernel_size=3, stride=1, padding=1)
        self.conv2_2 = nn.Conv2d(int(decoder_outdim*8/4), int(decoder_outdim*8/4), kernel_size=3, stride=1, padding=1)
        self.bn2_1 = nn.BatchNorm2d(int(decoder_outdim*8/4))
        self.bn2_2 = nn.BatchNorm2d(int(decoder_outdim*8/4))
        
        self.upconv3 = nn.ConvTranspose2d(int(decoder_outdim*8/4), int(decoder_outdim*8/8), kernel_size=2, stride=2)
        # self.upconv3 = nn.ConvTranspose2d(int(decoder_outdim*8/4), int(decoder_outdim*8/8), kernel_size=2, stride=2, output_padding=1)
        self.bn3 = nn.BatchNorm2d(int(decoder_outdim*8/8))
        self.conv3_1 = nn.Conv2d(int(decoder_outdim*8/8), int(decoder_outdim*8/8), kernel_size=3, stride=1, padding=1)
        self.conv3_2 = nn.Conv2d(int(decoder_outdim*8/8), int(decoder_outdim*8/8), kernel_size=3, stride=1, padding=1)
        self.bn3_1 = nn.BatchNorm2d(int(decoder_outdim*8/8))
        self.bn3_2 = nn.BatchNorm2d(int(decoder_outdim*8/8))
        
        self.conv1x1 = nn.Conv2d(int(decoder_outdim*8/8), 1, kernel_size=1, stride=1, padding=0)
    
    def forward(self, x):
        x = self.upconv1(x)
        print("upconv1 output shape:", x.shape)
        x = self.bn1(x)
        x = self.conv1_1(x)
        print("conv1_1 output shape:", x.shape)
        x = self.conv1_2(x)
        print("conv1_2 output shape:", x.shape)
        x = self.bn1_1(x)
        x = self.bn1_2(x)
        
        x = self.upconv2(x)
        print("upconv2 output shape:", x.shape)
        x = self.bn2(x)
        x = self.conv2_1(x)
        print("conv2_1 output shape:", x.shape)
        x = self.conv2_2(x)
        print("conv2_2 output shape:", x.shape)
        x = self.bn2_1(x)
        x = self.bn2_2(x)
        
        x = self.upconv3(x)
        print("upconv3 output shape:", x.shape)
        x = self.bn3(x)
        x = self.conv3_1(x)
        print("conv3_1 output shape:", x.shape)
        x = self.conv3_2(x)
        print("conv3_2 output shape:", x.shape)
        x = self.bn3_1(x)
        x = self.bn3_2(x)
        
        x = self.conv1x1(x)
        print("conv1x1 output shape:", x.shape)
        
        return x

# 示例输入
decoder_outdim = 12  # 你可以根据需求修改
model = YourModel(decoder_outdim)
input_tensor = torch.randn(1, decoder_outdim * 8, 45, 90)  # 假设输入尺寸为 [1, 64, 45, 90]
output = model(input_tensor)
```
