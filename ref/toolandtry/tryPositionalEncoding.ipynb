```python
import numpy as np

def log_scale_transform(min_value, max_value, num_points):
    # 计算对数坐标系下的最小值和最大值
    log_min = np.log10(min_value)
    log_max = np.log10(max_value)

    # 在对数坐标系下均匀取点
    log_points = np.linspace(log_min, log_max, num_points)

    # 将对数坐标系下的点转换回线性坐标系
    points = np.power(10, log_points)

    return points

# 设置原始范围和离散点数量
min_value = 0.001
max_value = 8
num_points = 280

# 在对数坐标系下放大和压缩值
transformed_points = log_scale_transform(min_value, max_value, num_points)

# 打印转换后的离散点
for point in transformed_points:
    print(point)
```

```python
import numpy as np

def transform_to_log_coordinates(x):
    x = np.where(x <= 0, np.finfo(float).eps, x)    # 将负值和零转换为正值
    log_x = np.log10(x)    # 取对数
    log_x = (log_x - np.min(log_x)) / (np.max(log_x) - np.min(log_x))    # 归一化处理
    return log_x

# 示例
x = np.array([0.001, 0.002, 0.01, 0.1, 0.5, 1, 8])  # 普通坐标
log_x = transform_to_log_coordinates(x)  # 转换为对数坐标

print(log_x)
```

```python
import torch

def transform_to_log_coordinates(x):
    log_x = (torch.log10(torch.where(x <= 0, torch.finfo(torch.float).eps, x)) - torch.log10(torch.tensor(0.001))) / (torch.log10(torch.tensor(8)) - torch.log10(torch.tensor(0.001)))
    return log_x

# 示例
x = torch.tensor([0.001, 0.002, 0.5, 1, 10, 100, 1000])  # 普通坐标
log_x = transform_to_log_coordinates(x)  # 转换为对数坐标

print(log_x)
```
