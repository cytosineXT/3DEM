```python
import time
import torch
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize

def plot2DRCS(rcs, cutmax=None):
    tic = time.time()
    print(rcs.shape)
    vmin = torch.min(rcs)
    vmax = torch.max(rcs)
    norm = Normalize(vmin=vmin, vmax=vmax)
    cmap = cm.jet

    plt.figure()
    plt.imshow(rcs, cmap=cmap, norm=norm, origin='lower')
    plt.colorbar(label='RCS/m')
    plt.xlabel("Theta")
    plt.ylabel("Phi")
    if cutmax != None:
        plt.ylim(0, cutmax) 
    plt.show()
    print(f'{time.time()-tic:.4f}s')

rcs = torch.load('./ref/gt.pt')[0,:,:]
plot2DRCS(rcs)
print(rcs)

rcs_log10 = torch.log10(rcs)
plot2DRCS(rcs_log10)
print(rcs_log10)

rcs_original = torch.pow(10, rcs_log10)
plot2DRCS(rcs_original)
print(rcs_original)

print("Original RCS shape:", rcs.shape)
print("Log10 transformed RCS shape:", rcs_log10.shape)
print("Reverse transformed RCS shape:", rcs_original.shape)

print(rcs == rcs_original)
print(torch.allclose(rcs, rcs_original))
```

```python
import time
import torch
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize

def plot2DRCS(rcs, cutmax=None):
    tic = time.time()
    print(rcs.shape)
    vmin = torch.min(rcs)
    vmax = torch.max(rcs)
    norm = Normalize(vmin=vmin, vmax=vmax)
    cmap = cm.jet

    plt.figure()
    plt.imshow(rcs, cmap=cmap, norm=norm, origin='lower')
    plt.colorbar(label='RCS/m')
    plt.xlabel("Theta")
    plt.ylabel("Phi")
    if cutmax != None:
        plt.ylim(0, cutmax) 
    plt.show()
    print(f'{time.time()-tic:.4f}s')


rcs = torch.load('./ref/gt.pt')[0,:,:]
plot2DRCS(rcs)
print(rcs)

epsilon = 0.001
rcs_log10 = torch.log10(torch.max(rcs, torch.tensor(epsilon, device=rcs.device)))
plot2DRCS(rcs_log10)
print(rcs_log10)

rcs_original = torch.pow(10, rcs_log10)
plot2DRCS(rcs_original)
print(rcs_original)

print("Original RCS shape:", rcs.shape)
print("Log10 transformed RCS shape:", rcs_log10.shape)
print("Reverse transformed RCS shape:", rcs_original.shape)

print(rcs == rcs_original)
print(torch.allclose(rcs, rcs_original))
```
