```python
import time
import torch
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize

def plot2DRCS(rcs, savedir,logger,cutmax):
    tic = time.time()
    print(rcs.shape)
    vmin = torch.min(rcs)
    vmax = torch.max(rcs)
    norm = Normalize(vmin=vmin, vmax=vmax)
    cmap = cm.jet

    plt.figure()
    plt.imshow(rcs, cmap=cmap, norm=norm, origin='lower')
    plt.colorbar(label='RCS/m²')
    plt.xlabel("Theta")
    plt.ylabel("Phi")
    if cutmax != None:
        plt.ylim(0, cutmax) 
    plt.savefig(savedir)
    print(f'画图用时：{time.time()-tic:.4f}s')

outrcs = torch.load('decoded.pt')[0,:,:]
gt = torch.load('gt.pt')[0,:,:]
out2Drcspngpath = './2D.png'
out2Drcspngpath2 = './2D2.png'
gt2Drcspngpath = './2Dgt.png'
plot2DRCS(rcs=gt, savedir=gt2Drcspngpath, logger=None,cutmax=None)
plot2DRCS(rcs=outrcs, savedir=out2Drcspngpath, logger=None,cutmax=None)
print(torch.max(gt).item())
plot2DRCS(rcs=outrcs, savedir=out2Drcspngpath2, logger=None,cutmax=torch.max(gt).item())
```
