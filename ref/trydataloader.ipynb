# dataset

```python
import torch.utils.data.dataset as Dataset
import numpy as np
import torch
import os
import re
```

```python
rcsdir = r'/mnt/f/datasets/RCS_map'
in_ems = []
rcss = []
for file in os.listdir(rcsdir):
    # print(file)
    theta, phi, freq= re.search(r"RCSmap_theta(\d+)phi(\d+)f(\d.+).pt", file).groups()
    theta = int(theta)
    phi = int(phi)
    freq = float(freq)
    in_em = [theta,phi,freq]
    # print(in_em)
    rcs = torch.load(os.path.join(rcsdir,file))

    in_ems.append(in_em)
    rcss.append(rcs)
```

```python
print(in_ems)
print(rcss)
```

```python
#
class meshRCSDataset(Dataset.Dataset):
    #
    def __init__(self, Mesh, RCSmap):
        self.meshdata = Mesh
        self.RCSmap = RCSmap
    #
    def __len__(self):
        return len(self.meshdata)
    #
    def __getitem__(self, index):
        meshdata = torch.Tensor(self.meshdata[index])
        RCSmap = torch.Tensor(self.RCSmap[index])
        return meshdata, RCSmap
```

```python
dataset = meshRCSDataset(in_ems, rcss)

print(dataset)
print('dataset', dataset.__len__())
# print(dataset[0])
oneem, onercs = dataset[0]
print(oneem, onercs)
```

# dataloader

```python
import torch.utils.data.dataloader as DataLoader
```

```python
dataloader = DataLoader.DataLoader(dataset, batch_size=3, shuffle=True, num_workers=4) #DataLoader

for step, (data, label) in enumerate(dataloader):
    print(f'step{step} : data={data}, label=0')

# for i, item in enumerate(dataloader):
#     print('i:', i)
#     data, label = item
#     print('data:', data)
#     print('label:', label)

```
