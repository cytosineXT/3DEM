```python
import os

# dest_folder = "/home/jiangxiaotian/workspace/3DEM"
dest_folder = "/mnt/truenas_jiangxiaotian/allplanes/puredatasets"

print(f"目标文件夹{dest_folder}中的文件数量：{len(os.listdir(dest_folder))}")

total_size = 0

for root, dirs, files in os.walk(dest_folder):
    if root=='/mnt/truenas_jiangxiaotian/allplanes/puredatasets':
        continue
    '''
    root /mnt/truenas_jiangxiaotian/allplanes/puredatasets/b7fd11d4af74b4ffddaa0161e9d3dfac
    dirs
    files 是文件夹root下的文件list
    '''
    dir_size = 0
    file_count = len(files)  # 统计当前文件夹下的文件数量

    for file_name in files:
        file_path = os.path.join(root, file_name)
        file_size = os.path.getsize(file_path)
        dir_size += file_size
        total_size += file_size
    
    # 获取当前文件夹名（去掉完整路径）
    current_folder_name = os.path.basename(root)
    
    # 计算当前目录的大小并立即打印
    size_mb = dir_size / (1024 * 1024)  # 转换为 MB
    size_gb = dir_size / (1024 * 1024 * 1024)  # 转换为 GB
    print(f"子目录 {current_folder_name} 大小：{size_gb:.2f} GB, 文件数量：{file_count}")

# 打印总大小
total_size_mb = total_size / (1024 * 1024)  # 转换为 MB
total_size_gb = total_size / (1024 * 1024 * 1024)  # 转换为 GB
print(f"目标文件夹总大小：{total_size_gb:.2f} GB")
```

```python
import os
import shutil
import re
import time

starttime = time.time()

all_folder = '/mnt/truenas_jiangxiaotian/allplanes/puredatasets'
momdest_folder = "/mnt/truenas_jiangxiaotian/allplanes/mie"

for src_folder, dirs, files in os.walk(all_folder):
    tic = time.time()
    if src_folder=='/mnt/truenas_jiangxiaotian/allplanes/puredatasets':
        continue
    currentfoldername = os.path.basename(src_folder)
    pattern = r"RCSmap_theta(\d+)phi(\d+)f(\d+\.\d+).pt"
    plane = currentfoldername[:4]
    traindest_folder = os.path.join(momdest_folder, plane+'_mie_train')#/mnt/truenas_jiangxiaotian/allplanes/mie/baa9_mie_train
    os.makedirs(traindest_folder, exist_ok=True)

    # print(traindest_folder)
    # valdest_folder = traindest_folder.replace('_mie_train','_mie_val')
    # print(valdest_folder)
    # break

    for file_name in files:
        match = re.match(pattern, file_name)
        if match:
            theta = int(match.group(1))
            phi = int(match.group(2))
            f = float(match.group(3))
            if f >= 0.1 and f <= 1.0:
                src_file_path = os.path.join(src_folder, file_name)
                dest_file_path = os.path.join(traindest_folder, file_name.replace('RCSmap', plane))
                shutil.copy(src_file_path, dest_file_path)

    total_size = 0
    for root, dirs, files in os.walk(traindest_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            total_size += os.path.getsize(file_path)
    total_size_gb = total_size / (1024 * 1024 * 1024)  # 转换为 GB
    initrainlen = len(os.listdir(traindest_folder))
    initraingb = total_size_gb


    # #从train移成val
    valdest_folder = traindest_folder.replace('_mie_train','_mie_val')
    pattern = r"([a-zA-Z0-9]{4})_theta(\d+)phi(\d+)f(\d+\.\d+).pt"
    cnt = 940 #抽出1/10做验证集
    os.makedirs(valdest_folder, exist_ok=True)

    files = os.listdir(traindest_folder)
    for file_name in files:
        match = re.match(pattern, file_name)
        if match:
            plane = match.group(1)
            theta = int(match.group(2))
            phi = int(match.group(3))
            f = float(match.group(4))

            src_file_path = os.path.join(traindest_folder, file_name)
            dest_file_path = os.path.join(valdest_folder, file_name)
            shutil.move(src_file_path, dest_file_path)
            cnt -= 1
        if cnt == 0:
            break

    total_size = 0
    for root, dirs, files in os.walk(valdest_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            total_size += os.path.getsize(file_path)
    total_size_mb = total_size / (1024 * 1024)  # 转换为 MB
    total_size_gb = total_size / (1024 * 1024 * 1024)  # 转换为 GB
    vallen = len(os.listdir(valdest_folder))
    valgb = total_size_gb


    #从train拷贝成pretrain
    predest_folder = traindest_folder.replace('_mie_train','_mie_10train')
    pattern = r"([a-zA-Z0-9]{4})_theta(\d+)phi(\d+)f(\d+\.\d+).pt"
    cnt = 940 #复制出1/10做pretrain
    os.makedirs(predest_folder, exist_ok=True)

    files = os.listdir(traindest_folder)
    for file_name in files:
        match = re.match(pattern, file_name)
        if match:
            plane = match.group(1)
            theta = int(match.group(2))
            phi = int(match.group(3))
            f = float(match.group(4))

            src_file_path = os.path.join(traindest_folder, file_name)
            dest_file_path = os.path.join(predest_folder, file_name)
            shutil.copy(src_file_path, dest_file_path)
            cnt -= 1
        if cnt == 0:
            break

    total_size = 0
    for root, dirs, files in os.walk(predest_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            total_size += os.path.getsize(file_path)
    total_size_mb = total_size / (1024 * 1024)  # 转换为 MB
    total_size_gb = total_size / (1024 * 1024 * 1024)  # 转换为 GB
    pre10len = len(os.listdir(predest_folder))
    pre10gb = total_size_gb
    

    #从train拷贝成pretrain50
    pre50dest_folder = traindest_folder.replace('_mie_train','_mie_50train')
    pattern = r"([a-zA-Z0-9]{4})_theta(\d+)phi(\d+)f(\d+\.\d+).pt"
    cnt = 940*5 #复制出1/10做pretrain
    os.makedirs(pre50dest_folder, exist_ok=True)

    files = os.listdir(traindest_folder)
    for file_name in files:
        match = re.match(pattern, file_name)
        if match:
            plane = match.group(1)
            theta = int(match.group(2))
            phi = int(match.group(3))
            f = float(match.group(4))

            src_file_path = os.path.join(traindest_folder, file_name)
            dest_file_path = os.path.join(pre50dest_folder, file_name)
            shutil.copy(src_file_path, dest_file_path)
            cnt -= 1
        if cnt == 0:
            break

    total_size = 0
    for root, dirs, files in os.walk(pre50dest_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            total_size += os.path.getsize(file_path)
    total_size_mb = total_size / (1024 * 1024)  # 转换为 MB
    total_size_gb = total_size / (1024 * 1024 * 1024)  # 转换为 GB
    pre50len = len(os.listdir(pre50dest_folder))
    pre50gb = total_size_gb


    for root, dirs, files in os.walk(traindest_folder):
        for file_name in files:
            file_path = os.path.join(root, file_name)
            total_size += os.path.getsize(file_path)
    finaltrainlen = len(os.listdir(traindest_folder))
    finaltraingb = total_size_gb

    print(f'原始文件夹{src_folder}，初始train {initrainlen} {initraingb:.2f}GB，val {vallen} {valgb:.2f}GB，10train {pre10len} {pre10gb:.2f}GB，50train {pre50len} {pre50gb:.2f}GB，最终train {finaltrainlen} {finaltraingb:.2f}GB，耗时{time.strftime("%H:%M:%S", time.gmtime(time.time()-tic))}')#草忘了final train

print(f'总耗时{time.strftime("%H:%M:%S", time.gmtime(time.time()-starttime))}')
```
